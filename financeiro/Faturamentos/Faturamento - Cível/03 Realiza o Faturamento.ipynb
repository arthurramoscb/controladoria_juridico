```python
import os
import shutil
import pandas as pd
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows

print("--- Iniciando o processo de salvamento por escritório ---")

# --- 1. Configuração dos Caminhos ---
# O caminho do template continua o mesmo
caminho_template = "/Volumes/databox/juridico_comum/arquivos/financeiro/Faturamento_Civel/Base_Gerencial_Modelo/Faturamento Mensal Cível - Novo Contrato.xlsx"

# ▼▼▼ CAMINHO DE SAÍDA ATUALIZADO CONFORME SUA SOLICITAÇÃO ▼▼▼
diretorio_saida_por_escritorio = f"/Volumes/databox/juridico_comum/arquivos/financeiro/Faturamento_Civel/FINAL/Escritorios/"

# Garante que o diretório de saída exista. Se não existir, ele o cria.
os.makedirs(diretorio_saida_por_escritorio, exist_ok=True)
print(f"Arquivos serão salvos em: {diretorio_saida_por_escritorio}")

# O DataFrame final e completo que será filtrado
df_completo = df_mod

# --- 2. Obter a lista de escritórios únicos ---
lista_escritorios = df_completo['ESCRITÓRIO EXTERNO'].dropna().unique().tolist()
print(f"Encontrados {len(lista_escritorios)} escritórios únicos para processar.")

# --- 3. Loop principal: um arquivo para cada escritório ---
for escritorio in lista_escritorios:
    print(f"\n--- Processando escritório: '{escritorio}' ---")

    # Sanitização do nome para o arquivo
    nome_arquivo_seguro = "".join(c for c in escritorio if c.isalnum() or c in (' ', '_')).rstrip()
    nome_arquivo_seguro = nome_arquivo_seguro.replace(' ', '_')

    # Filtragem dos dados
    df_filtrado = df_completo[df_completo['ESCRITÓRIO EXTERNO'] == escritorio]
    print(f"Encontradas {len(df_filtrado)} linhas para este escritório.")

    # Definição dos caminhos temporário e final
    caminho_temporario_escritorio = f"/tmp/{nome_arquivo_seguro}_{Mes}.xlsx"
    caminho_final_escritorio = os.path.join(diretorio_saida_por_escritorio, f"Consolidado_{nome_arquivo_seguro}_{Mes}.xlsx")

    # Lógica de injeção de dados no template
    try:
        shutil.copy(caminho_template, caminho_temporario_escritorio)
        workbook = load_workbook(caminho_temporario_escritorio)
        sheet_name = 'Consolidado FATURAMENTO'
        worksheet = workbook[sheet_name]
        
        rows = dataframe_to_rows(df_filtrado, index=False, header=False)
        
        for r_idx, row in enumerate(rows, 3):
            for c_idx, value in enumerate(row, 1):
                worksheet.cell(row=r_idx, column=c_idx, value=value)
        
        workbook.save(caminho_temporario_escritorio)
        shutil.move(caminho_temporario_escritorio, caminho_final_escritorio)
        print(f"✅ Arquivo salvo com sucesso em: {caminho_final_escritorio}")

    except Exception as e:
        print(f"❌ Ocorreu um erro ao processar o escritório '{escritorio}': {e}")
    finally:
        if os.path.exists(caminho_temporario_escritorio):
            os.remove(caminho_temporario_escritorio)

print("\n--- Processo de separação por escritório concluído! ---")
```
