```python
def normalizar_cabecalho(df_input):
    """
    Renomeia todas as colunas de um DataFrame (Pandas ou Spark):
    - Remove acentos
    - Substitui espaços e caracteres especiais por _
    - Converte para minúsculas
    """
    # Se for Spark DataFrame
    if isinstance(df_input, DataFrame):
        novas_colunas = []
        for col in df_input.columns:
            # Normalização string comum
            c_norm = unidecode(col) # Remove acentos
            c_norm = c_norm.lower() # Minúsculo
            c_norm = re.sub(r'[^a-z0-9]', '_', c_norm) # Tudo que não for letra/num vira _
            c_norm = re.sub(r'_+', '_', c_norm) # Remove underlines duplicados
            c_norm = c_norm.strip('_') # Remove underline do começo/fim
            novas_colunas.append(c_norm)
        
        return df_input.toDF(*novas_colunas)
    
    # Se for Pandas DataFrame
    elif isinstance(df_input, pd.DataFrame):
        df_input.columns = [
            re.sub(r'_+', '_', re.sub(r'[^a-z0-9]', '_', unidecode(c).lower())).strip('_')
            for c in df_input.columns
        ]
        return df_input
    
    else:
        print("❌ Tipo de DataFrame não suportado.")
        return df_input
```
